## Rumus Entropy




Terdapat 10 Yes dan 4 No, jika entry pada tabel dijumlahkan menjadi 14

## 1. Entropy(S)

```python
import math

S = -(10/14) * math.log2(10/14) - (4/14) * math.log2(4/14)
S
```

Nilai entropy yang bisa didapat adalah 0.863120568566631

## 2. Mencari Information Gain dari Outlook
Outlook memiliki 3 atribut:
- Sunny
- Cloudy
- Rainy

Jadi saya akan menghitung nilai entropy dari setiap atribut di atas.

a. Atribut Sunny ada 5 atribut dengan:
- 2 Yes (Play: Play)
- 3 No (Play: Don't Play)

```python
entropy_S_sunny = -(2/5) * math.log2(2/5) - (3/5) * math.log2(3/5)
entropy_S_sunny
```

b. Atribut Cloudy ada 4 atribut dengan:

- 4 Yes 
- 0 No

```python
entropy_S_cloudy = -(4/4) * math.log2(4/4) 
entropy_S_cloudy
```

c. Atribut Rainy ada 5 atribut dengan:

- 4 yes
- 1 No

```python
entropy_S_rainy = -(4/5) * math.log2(4/5) - (1/5) * math.log2(1/5)
entropy_S_rainy
```

Sudah ada 4 Entropy:
- Entropy(S) = 0.86
- Entropy(S-Sunny) = 0.97
- Entropy(S-Cloudy) = 0
- Entropy(S-Rainy) = 0.72

Rumus Information Gain


```python
gain_outlook = 0.86-5/14*(0.97)-4/14*(0)-5/14*(0.72)
gain_outlook
```
